```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('default')
```

# Sinais Cont√≠nuos e Discretos

```python
# Sinal Continuo
t = np.linspace(0, 20, 101)
y_con = np.cos(t) * np.e**(-0.1*t)

# Sinal Discreto
n = np.linspace(0, 20, 21)
y_dis = np.cos(n) * np.e**(-0.1*n)

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_con);plt.ylim(-1.5, 1.5)
plt.subplot(1,2,2);plt.stem(n, y_dis);plt.plot(t, y_con, '--');plt.ylim(-1.5, 1.5)
```

# Amostragem de Sinais

```python
f_s = 0.5     # C√°lculo em Hz
T_s = 1/f_s   # C√°lculado em s

n = np.linspace(0, 40, 41)
y_sampled = np.cos(n*T_s)*np.power(np.e, -0.1*n*T_s)

plt.figure(figsize = (10,5))
plt.stem(n/f_s, y_sampled);plt.ylim(-1.5, 1.5);
```

# Discretiza√ß√£o de um sinal cont√≠nuo

```python
# sinal continuo
f0 = 3
t = np.arange(0,1,0.0002)
xc = np.cos(f0*t*2*np.pi)

#Discretiza√ß√£o do sinal 
fs = 6
Ts = 1/fs
n = np.linspace(0,1*fs, int(1*fs)+1)
xd = np.cos(f0*2*np.pi*n*Ts)

plt.figure(figsize = (10,5));
plt.stem(n*Ts,xd, '--r');plt.plot(t,xc);
```

# Sinal Par

```python
plt.rcParams['figure.figsize'] = [12, 5]

t = np.linspace(-10, 10, 101)
y_parc = np.cos(t)*np.exp(-0.1*np.abs(t))

n = np.linspace(-10, 10, 21)
y_pard = np.cos(n)*np.exp(-0.1*np.abs(n))

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_parc)
plt.subplot(1,2,2);plt.stem(n, y_pard);plt.plot(t, y_parc, '--')
```

# Sinal √çmpar

```python
t = np.linspace(-10, 10, 101)
y_impc = np.sin(t)*np.exp(-0.1*np.abs(t))

n = np.linspace(-10, 10, 21)
y_impd = np.sin(n)*np.exp(-0.1*np.abs(n))

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_impc)
plt.subplot(1,2,2);plt.stem(n, y_impd);plt.plot(t, y_impc, '--')
```

# Decomposi√ß√£o Par/Impar de um sinal composto

```python
#sinal continuo
t = np.arange(-6,6,0.0002)
x = np.zeros(t.shape)

x[(t <=-5) & (t >= 5)] = 0
x[(t > -5) & (t <=-1)] = 1
x[(t > -1) & (t <= 1)] = 3
x[(t >  1) & (t <  5)] = 5

#parte par do sinal
xp = (x + x[::-1])/2

#parte impar do sinal 
xi = (x - x[::-1])/2

#soma das partes par e impar do meu sinal 
xt = xp + xi

plt.figure(1);plt.plot(t, x, drawstyle = 'steps-pre');plt.title('Sinal continuo')

plt.figure(2, figsize = (15,5) );
plt.subplot(1,2,1);plt.plot(t, xp,'g', drawstyle = 'steps-pre');plt.title('parte par do sinal ')
plt.subplot(1,2,2);plt.plot(t, xi,'r', drawstyle = 'steps-pre');plt.title('parte impar do sinal')
plt.figure(3);plt.plot(t, xp,'--g', drawstyle = 'steps-pre');plt.plot(t, xi,'--r', drawstyle = 'steps-pre');plt.plot(t, xt,'black', drawstyle = 'steps-pre');plt.title('soma da parte PAR + IMPAR')
plt.figure(4);plt.plot(t, xt,'black', drawstyle = 'steps-pre');plt.title('soma da parte PAR + IMPAR')
```

# Transforma√ß√µes Lineares da Vari√°vel Independente

# Invers√£o

```python
t = np.linspace(-10, 10, 101)
y_original = np.cos(t)*np.exp(-0.1*t)
y_inversa = np.cos(-t)*np.exp(-0.1*-t)

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_original)
plt.subplot(1,2,2);plt.plot(t, y_inversa)
```

# Compress√£o

```python
a = 2

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_compressao = np.sin(a*t)*np.exp(-0.1*np.abs(a*t))

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_original)
plt.subplot(1,2,2);plt.plot(t, y_compressao)
```

# Expans√£o

```python
a = 0.5

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_expansao = np.sin(a*t)*np.exp(-0.1*np.abs(a*t))

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_original)
plt.subplot(1,2,2);plt.plot(t, y_expansao)
```

# Avan√ßo

```python
b = 4

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_avanco = np.sin(t+b)*np.exp(-0.1*np.abs(t+4))

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_original);plt.title('Original')
plt.subplot(1,2,2);plt.plot(t, y_avanco);plt.title('Avan√ßado')
```

# Atraso

```python
b = 4

t = np.linspace(-10, 10, 101)
y_original = np.sin(t)*np.exp(-0.1*np.abs(t))
y_atraso = np.sin(t-b)*np.exp(-0.1*np.abs(t-b))

plt.figure(figsize = (15,5))
plt.subplot(1,2,1);plt.plot(t, y_original)
plt.subplot(1,2,2);plt.plot(t, y_atraso);
```
