```python
from IPython.display import Audio
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Dicionrio musical
framerate = 44100

time = 0.5
t = np.linspace(0, time, int(framerate*time))

notas_dic = {'C': 261.63,
             'C#': 277.18,
             'D': 293.66,
             'D#': 311.13,
             'E': 329.63,
             'F': 349.23,
             'F#': 369.99,
             'G': 392.00,
             'G#': 415.30,
             'A': 440.00,
             'Bb': 466.16,
             'B': 493.88,
             'D1': 587.33}

# Melodia da msica "Samurai" do Djavan
melodia = ['C', 'C', 'D', 'C', 'F', 'E',
         'C', 'C', 'D', 'C', 'G', 'F',
         'C', 'C', 'B', 'A', 'F', 'E', 'C#',
         'A', 'A', 'G','D', 'G', 'D']

# Gerao dos sinais sonoros
sons = np.zeros((len(melodia), int(framerate*time)))

for i in range(len(melodia)):
    sons[i] = np.sin(2*np.pi*notas_dic[melodia[i]]*t)

# Reproduo da melodia
Audio(np.concatenate(sons), rate=framerate)
```
