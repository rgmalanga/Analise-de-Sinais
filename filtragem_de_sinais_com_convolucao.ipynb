# Exercício

Escreva um código construindo um sinal discreto $A$, corrompido por ruído branco, e um sinal $B$ que será um filtro, e realize a convolução destes dois sinais de forma a eliminar os problemas do sinal através do processo de convolução. Represente graficamente os três sinais obtidos no processo. Abaixo segue a descrição dos sinais:
    
  \begin{equation} 
  A[n] = \sin[2 \cdot \pi \cdot \frac{1}{50} \cdot n] + \mathcal{N}(\sigma = 0.1, \mu = 0) \quad 0 \leq n \leq 50
  \end{equation}
  
  \begin{equation} 
  B[n] = \frac{1}{11} \quad 0 \leq n \leq 10
  \end{equation}

```python
import numpy as np
from matplotlib import pyplot as plt 


n = np.arange(0,51,1)

#ruido branco 
sigma = 0.1; mi = 0
ruido = np.random.randn(n.shape[0])*sigma + mi 

#sinal discreto A

A = np.sin(2*np.pi*(1/50)*n) + ruido

#sinal FILTRO
nB = np.linspace(0,10,11)
B = [1/11] *11

#convolução 

conv = np.convolve(A,B)

plt.figure(figsize = (12,3));plt.tight_layout();
plt.subplot(1,3,1);plt.plot(n,A);plt.title('sinal corrompido')
plt.subplot(1,3,2);plt.plot(nB,B);plt.title('FILTRO')
plt.subplot(1,3,3);plt.plot(conv);plt.title('sinal filtrado');
```
